<h3>Mean and median imputation</h3>
<h4>1.what is imputation?</h4>
    <p>-Imputation is the act of replacing missing data with statistical estimates of the missing values.<p>
<h4>2.Mean imputation:</h4>
    <p>-Mean imputation consists of replacing all occurrences of missing values (NA) within a variable
    by the mean (if the variable has a Gaussian(Normal) distribution
    [Gaussian Distribution which typically has two parameters: mean and standard deviation (SD)])<p>
<h4>3.Median imputation:</h4>
   <p>- median imputation consists of replacing all occurrences of missing values (NA) within a variable
    by the median (if the variable has a skewed distribution).<p>
    
   <p> --Mean: The mean is also known as the average, 
            and it is calculated by adding up all the values in a data set and dividing by the total number of values.<p>
    <p>--Median:The median is the middle value of a data set,
              which separates the highest and lowest values equally. It is calculated by arranging the data set in order from lowest to highest and finding the value in the exact middle.<p>
   <p> --Mode: The mode is the value that appears most frequently in a data set(মোড হল সেই মান যা একটি ডেটা সেটে প্রায়শই দেখা যায়)<p>

```python
import pandas as pd
import numpy as np
```

```python
#DataFrame
m=pd.read_excel('coalpublic2013.xlsx')
```

```python
m
```

```python
m.dropna() # the 'Dropna()' method will be returned to which rows are no missing values.
```

```python
'''
    Mean imputation is often used when the missing values are numerical and the distribution of 
    the variable is approximately normal.
'''
m.isnull().mean() #The mean is also known as the average
```

```python
m.fillna(m.mean())
```

```python
'''
Impute / Replace Missing Values with Median
1.When use the Median imputaion? 
     -Median imputation is preferred when the distribution is skewed, 
      as the median is less sensitive to outliers than the mean.
Skewed: A skewed distribution occurs when one tail is longer than the other.
'''
#Note that imputing missing data with mean values can only be done with numerical data.
m.fillna(m.median())
```

```python
'''
Impute / Replace Missing Values with Mode
1.when use the mode imputaion?
    --Mode imputation is suitable for categorical variables or numerical variables 
      with a small number of unique values.
'''

m.fillna(m['Year2'].mode()[0])
```

```python
#b)Arbitrary value imputaion
'''
1.Which variables can I impute with an arbitrary value?
    -Both categorical and numerical variables can be imputed by arbitrary values.
    
    Arbitrary value imputation consists of replacing all occurrences of missing values within a variable 
    with an arbitrary value. Typically used arbitrary values are 0, 999, -999 (or other combinations of 9s)
    or -1 (if the distribution is positive)
'''
m['Year2'].fillna(value = 2026)
```

```python
# c) Frequent category imputation
'''
Frequent category imputation:
    -This technique says to replace the missing value with the variable with the highest frequency or in 
     simple words replacing the values with the Mode of that column.This technique is also referred to as
     Mode Imputation.
'''
#একটা কলাম এ যে ভ্যালু  বেশি থাকে সেটাই মোড[2016=4]
m['Year2'].fillna(m['Year2'].mode()[0])
```

```python
#d) Missing category imputation
'''
Missing category imputaion:
    -Missing category imputaion is a technique used to replace the missing values in a categorical variable
    with a new category('unknown')
'''
m.fillna(value='unknown')
```
